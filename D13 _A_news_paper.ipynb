Fitting a Linear Regression Model

```python
import statsmodels.formula.api as smf
model = smf.ols("sunday~daily",data = data).fit()


# statsmodel library   ------>   Module powerful Pyt lib used for statistical modeling and testing
# .formula.api         ------>   sub model using
# sunday               ------>   y
# daily                ------>   x variable
# ols---. Ordinary Least Squares (OLS) regression model. OLS is a common method for fitting linear regression models.
```

```python
model.summary()

"""   R-squared  , Degree of freedom Model(dependent var Y) ,Intercept """
```

```python
sns.regplot(x="daily", y="sunday", data=data);
```

```python
#Coefficients

model.params

#params ---> estimated coefficients of the linear regression model
```

```python
#t and p-Values
print(model.tvalues, '\n', model.pvalues)

"""  """
```

```python
#R squared values
(model.rsquared,model.rsquared_adj)
```

# Predict for new data point

```python
#Predict for 200 and 300 daily circulation
newdata=pd.Series([200,300])
```

```python
data_pred=pd.DataFrame(newdata,columns=['daily'])
```

```python
model.predict(data_pred)
```
